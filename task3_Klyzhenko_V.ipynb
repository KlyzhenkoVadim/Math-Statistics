**Решение:**

* $\widehat{\theta_1}, ... , \widehat{\theta_k}$ 

```python
def Theta(k,n):
    Sample=sps.norm.rvs(size=(k,n))
    Z=np.array([Sample[i,:].mean() for i in range(k)])
    A =np.array([np.sum((Sample[i,:]-Z[i])**2/n) for i in range(k)])
    return A
```

**Решение:**

* $\overline{X}$ - выборочное среднее:

```python
n=1000
sample = sps.cauchy.rvs(size=n)
```

```python
mean=sample.cumsum()/(np.arange(n)+1)  
plt.figure(figsize=(16,9))
plt.plot(mean,color='r',label=r'$\overline{X}$')
plt.title('Выборочное Среднее',fontsize=18)
plt.ylabel('Cреднее',fontsize=16)
plt.xlabel('n',fontsize=16)
plt.legend(fontsize=17)
plt.hlines(0,0,n)
```

* $\widehat{\mu}$ —  выборочная медиана;

```python
mediana=np.array([np.median(sample[:i]) for i in range(1,(n+1))])
grid = np.linspace(1, n, n)  
plt.figure(figsize=(16,9))
plt.plot(grid,mediana,color='blue',label=r'$\widehat{\mu}$')
plt.title('Выборочная Медиана',fontsize=20)
plt.ylabel(r'$\widehat{\mu}$',fontsize=20)
plt.xlabel('n',fontsize=16)
plt.legend(fontsize=20)
plt.hlines(0,0,n,color='g')
```

* Oдношаговая оценка, построенная по выборочной медиане (см. лекцию 4):

```python
theta1=np.array([mediana[i]+np.sum((sample[:i]-mediana[i])/(1+(sample[:i]-mediana[i])**2))/np.sum((1-(sample[:i]-mediana[i])**2)/(1+(sample[:i]-mediana[i])**2)**2) for i in range(n)])
grid = np.linspace(1, n, n)  
plt.figure(figsize=(16,9))
plt.plot(grid,theta1,color='blue')
plt.title('Одношаговая оценка',fontsize=20)
plt.ylabel(r'$\widehat{\mu_1}$',fontsize=20)
plt.xlabel('n',fontsize=16)
plt.hlines(0,0,n,color='g')
theta1.shape
```

**Вывод:** По графикам видно, что выборочное среднее плохо подходит для оценки параметра $\theta$, я думаю, из-за того, что в распределении Коши не существует матожидания, а по ЗБЧ среднее к нему стремится. Оценки при помощи медиан подходят наилучшим образом.

Реализуйте функции, возвращающие значения функции правдоподобия  и логарифмической функции правдоподобия для гамма распределения с параметрам $\theta, \beta$. Функция принимает реализацию выборки и параметры. Логарифмическую функцию реализуйте с использованием формулы после ее упрощения, а не как взятие функции логарифма от уже вычисленного значения функции правдоподобия.

Для получения частичного балла можно использовать методы `pdf` и `logpdf` для плотности и логарифма плотности распределения в `scipy.stats`, для получения полного балла функции формулу нужно реализовать самостоятельно.

Для подсчета гамма-функции в знаменателе плотности распределения используйте `scipy.special.gamma`, а для подсчета логарифма гамма-функции &mdash; `scipy.special.loggamma`. 



```python
import scipy.special
def L(sample,beta,theta):
    g = scipy.special.gamma(beta)**sample.size
    t = theta**(sample.size*beta)
    prod= np.prod(sample**(beta-1))
    ex=np.exp(-theta*(np.sum(sample)))
    L = t*prod*ex/g
    return L
L(df['ITSN1_N'],13.5,1)

def Ln(sample,beta,theta):
    n=sample.size
    L1 = n*beta*np.log(theta)+(beta-1)*np.sum(np.log(sample))-theta*np.sum(sample)-n*scipy.special.loggamma(beta)
    return L1
Ln(df['ITSN1_N'],13.5,theta)
```
